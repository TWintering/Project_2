```python
import json
import requests
import pandas as pd
import numpy
import pymongo
```

```python
electionURL = 'https://www.politico.com/election-results/2018/ohio/county.json'

response = requests.get(electionURL).json()
data = json.dumps(response, indent = 2)
elect_data = pd.read_json(data)
elect_data.head()
```

```python
type(data)
```

```python
fp=elect_data['fipscode'].tolist()
```

```python
fips=[]
for f in fp:
    if f not in fips:
        fips.append(f)
```

```python
gop_g=[]
gop_s=[]
gop_h=[]
dem_g=[]
dem_s=[]
dem_h=[]
tot_g=[]
tot_s=[]
tot_h=[]
share_g=[]
share_s=[]
```

```python
for f in fips:
    for r in elect_data.itertuples():
        if f==r[1]:
            if r[4]=="GOP":
                if r[3]=="G":
                    gop_g.append(r[14])
                if r[3]=="S":
                    gop_s.append(r[14])
            if r[4]=="Dem":
                if r[3]=="G":
                    dem_g.append(r[14])
                if r[3]=="S":
                    dem_s.append(r[14])
```

```python
dem_g_share=[]
gop_g_share=[]
dem_s_share=[]
gop_s_share=[]
```

```python
for i in range(len(fips)):
    tot_s.append(dem_s[i]+gop_s[i])
    tot_g.append(dem_g[i]+gop_g[i])
    dem_s_share.append(dem_s[i]/(dem_s[i]+gop_s[i]))
    dem_g_share.append(dem_g[i]/(dem_g[i]+gop_g[i]))
    gop_s_share.append(gop_s[i]/(dem_s[i]+gop_s[i]))
    gop_g_share.append(gop_g[i]/(dem_g[i]+gop_g[i]))
```

```python
len(tot_s)
```

```python
sen=pd.DataFrame({"fipscode":fips,"dem_votes":dem_s,"gop_votes":gop_s,"total_votes":tot_s,"dem_percent":dem_s_share,"gop_percent":gop_s_share})
```

```python
gov=pd.DataFrame({"fipscode":fips,"dem_votes":dem_g,"gop_votes":gop_g,"total_votes":tot_g,"dem_percent":dem_g_share,"gop_percent":gop_g_share})
```

```python
edu_csv = r'Education.csv'
edu_read = pd.read_csv(edu_csv, encoding='utf-8')

edu_read1 = edu_read.iloc[:,[0,1,2,43,44,45,46]]
edu_read2 = edu_read1.rename(columns= {"Area name":"County","FIPS Code":"fipscode",'Percent of adults with less than a high school diploma, 2012-2016':'less_than_HighSchool','Percent of adults with a high school diploma only, 2012-2016':"has_HighSchool","Percent of adults completing some college or associate's degree, 2012-2016":"some_College","Percent of adults with a bachelor's degree or higher, 2012-2016":"Undergrad_plus"}).set_index('fipscode')
edu_read3 = edu_read2.loc[edu_read2['State'] == 'OH']
edu_read3.head()
```

```python
senate = pd.merge(sen, edu_read3, how='left', on='fipscode')
senate.set_index("County")
```

```python
senate['dem_percent']=round((senate['dem_percent']*100),2)
senate['gop_percent']=round((senate['gop_percent']*100),2)
senate.head()
```

```python
governor = pd.merge(gov, edu_read3, how='left', on='fipscode')
governor.set_index("County")
```

```python
governor['dem_percent']=round((governor['dem_percent']*100),2)
governor['gop_percent']=round((governor['gop_percent']*100),2)
governor.head()
```
